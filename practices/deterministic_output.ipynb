# How To Get Deterministic Output

````python
import os
from langchain_groq import ChatGroq
from dotenv import load_dotenv
load_dotenv()

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")

model = ChatGroq(
    model='llama-3.1-8b-instant',
    # temperature=1,
    # max_tokens=10,
)
````

````python
response = model.invoke("What is the capital of France?")
response2 = model.invoke("Write a short motivational quote.")

print(response.content)
print(response2.content)
````

````python
user_info = {
    "Name": "Safi",
    "Age": 22
}

prompt = f"""
Given the following user information:

{user_info}

Provide the user's details.
"""

response = model.invoke(prompt)

print(response.content)
````

````python
user_info = {
    "Name": "Safi",
    "Age": 22
}

prompt = f"""
Given the following user information:

{user_info}

Provide the user's details.

Return ONLY valid JSON.
"""

response = model.invoke(prompt)

print(response.content)
````

````python
from pydantic import BaseModel

class UserInfo(BaseModel):
    name: str
    age: int

structured_model = model.with_structured_output(
    UserInfo
)

response = structured_model.invoke(
    "John is 25 years old."
)

print(response)
````

## Using Fixed Prompt

When you want the model to use data from a Python dictionary, you need to include that data in the prompt.

```python
user_info = {
    "Name": "John",
    "Age": 25
}

prompt = f"""
Given the following user information:

{user_info}

Provide the user's details.

Return ONLY valid JSON.
"""

response = model.invoke(prompt)

print(response.content)
```

````python
from langchain_core.tools import tool

@tool
def calculator(a: int, b: int):
    """
    Multiply two numbers.
    """
    return a * b

for i in range(5):
    print(
        calculator.invoke(
            {"a": 456, "b": 789}
        )
    )
````
